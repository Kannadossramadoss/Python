```python
# 3. Body Mass Index. Source:https://my.clevelandclinic.org/health/articles/9464-body-mass-index-bmi

print("Process is started")
while True:
    try:
        user_bmi = float(input("Enter the BMI index:"))
        if user_bmi <= 0:
            print("BMI must be greater than zero.")
            continue
        
        if user_bmi < 18.5:
            category = "Underweight"
        elif user_bmi < 25:
            category = "Optimum range"
        elif user_bmi < 30:
            category = "Overweight"
        else:
            category = "Very Overweight"
        print(category)
        break
    except ValueError:
        print("BMI of the user should be a number and greater than zero.")
print("Process is completed")
```
